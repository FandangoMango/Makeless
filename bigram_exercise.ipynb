```python
words = open('names.txt', 'r').read().splitlines()
```

```python
b = {}
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for bigram in zip(chs, chs[1:]):
        b[bigram] = b.get(bigram, 0) + 1
```

```python
sorted_b = sorted(b.items(), key = lambda kv: -kv[1])
```

```python
freq_dict = {}
for ((c1, c2), count) in sorted_b:
    if c1 not in freq_dict:
        freq_dict[c1] = {}
    freq_dict[c1][c2] = count
freq_dict
```

```python
from random import randint
word = ['<S>']
while word[-1] != '<E>':
    c = word[-1]
    keys = list(freq_dict[c].keys())
    vals = list(freq_dict[c].values())
    val_sum = sum(vals)
    rand_num = randint(0, val_sum -1)
    running_total = 0
    i = 0
    while running_total < rand_num:
        running_total += vals[i]
        i += 1
    word.append(keys[i])
word
```
